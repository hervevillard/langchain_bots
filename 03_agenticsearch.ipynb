```python
import os
from langchain_community.tools.tavily_search import TavilySearchResults
```

```python
client = TavilySearchResults(max_results=2)
client.name
```

```python
# run search

result = client.invoke({'query':"What is in Nvidia's new Blackwell GPU?"})
```

```python
print(result[0]['content'])
```

```python
# Choose a location
city = "Kigali"
query = {"query": f"""what is the current weather in {city}?
    Should I travel there today?
    "weather.com" """}
```

```python
import json
from pygments import highlight, lexers, formatters
```

```python
result = client.invoke(query)
```

```python
#parse json
parsed_json = json.loads(result[0]['content'].replace("'", '"'))

#pretty print JSON with syntaz higlighting
formatted_json = json.dumps(parsed_json, indent=4)
colorful_json = highlight(formatted_json,
                          lexers.JsonLexer(),
                          formatters.TerminalFormatter())

print(colorful_json)
```
